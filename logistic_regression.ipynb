```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


dataset = pd.read_csv('winequality-red.csv') #read the data
dataset.head()
```

```python
#outlier
X_out = np.random.rand(100, 5) * 100
y_out = X_out.dot(np.array([1, 2, 3, 4, 5])) + np.random.randn(100) * 10

# calculate the Z-score for each data point
z_scores = np.abs((y_out - np.mean(y_out)) / np.std(y_out))

# set a threshold for the Z-score
threshold = 3
# identify the outliers
outliers = np.where(z_scores > threshold)

# remove the outliers from the data
X_clean = np.delete(X_out, outliers, axis=0)
y_clean = np.delete(y_out, outliers)
```

```python
#setting the matrixes
X = dataset.iloc[:,0:11]
X_n=(X-X.mean())/X.std()
ones = np.ones([X_n.shape[0],1])
X_n = np.concatenate((ones,X_n),axis=1)

y = dataset.iloc[:,11].values #.values converts it from pandas.core.frame.DataFrame to numpy.ndarray


#set hyper parameters
alpha = 0.1
epochs = 3000

y
```

```python
y_he=[]
for i in y:
    if(i==3):
        y=[1,0,0,0,0,0]
    elif(i==4):
        y=[0,1,0,0,0,0]
    elif(i==5):
        y=[0,0,1,0,0,0]
    elif(i==6):
        y=[0,0,0,1,0,0]
    elif(i==7):
        y=[0,0,0,0,1,0]
    elif(i==8):
        y=[0,0,0,0,0,1]
    y_he.append(y)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split( X_n, y_he, test_size = 0.25,random_state=605)

w = np.zeros((6,12))
b = np.zeros(len(Ytrain))

```

```python
def Softmax(X_train,w):
    z=np.dot(X_train,w.T)
    sm=[]
    for i in range(len(z)):
        sm_temp=[]
        for j in range(len(z[i])):
            a=np.exp(z[i][j])/np.sum(np.exp(z[i]))
            sm_temp.append(a)
        sm.append(sm_temp)
    return sm
a=Softmax(Xtrain,w)

    
```

```python
def Loss(a,Y_train):
    loss=0
    for i in range(len(a)):
        for j in range(len(a[i])):
            loss+= Y_train[i][j]*np.log(a[i][j])
    return ((-1)/len(a))*(loss)
Loss(a,Ytrain)
```

```python
Ytrain = np.array(Ytrain)
Ytest = np.array(Ytest)
```

```python
def gradient_descent(x_train,y_train,w,alpha,epoch,a):
    cost=[]
    for iter in range(epoch):        
        total=alpha*np.dot(x_train.T,(y_train-a))/len(a)
        w=w+total.T
        a=Softmax(x_train,w)
        cost.append(Loss(a,y_train))
    return w,cost
w,c=gradient_descent(Xtrain,Ytrain,w,alpha,epochs,a)
c
```

```python
y_pred_prob = Softmax(Xtest, w)
y_pred = np.argmax(y_pred_prob, axis=1)
accuracy = np.mean(y_pred == np.argmax(Ytest, axis=1)) * 100
print("Accuracy:", accuracy, "%")
```
